```python
import pandas as pd
```

Lecturas del archivo
*Loading the dataset*

```python
df = pd.read_csv("Sales-Data-Analysis.csv")
```

```python
display(df.head())
df.info()
```

Organizamos el formato de la columna fecha para que no sea Object sino datetime64

*Converting the date column format from 'Object' to 'datetime64'*

```python
df['Date'] = pd.to_datetime(df['Date'], format='mixed')
```

```python
df.info()
```

Luego de revisar lo valores de cada una de las columnas notamos que en la columna 'Manager' habia que hacer una limpieza ya que nos registraban nombres repetidos los cuales pertenecian a una sola persona, esto sucedia ya que habian espacios al principio o mas de un espacio entre nombre y apellido.

Data Cleaning: 'Manager' column
After reviewing the values in each column, we noticed that the 'Manager' column required cleaning. There were duplicated records belonging to the same person due to formatting issues, such as leading spaces or multiple spaces between the first and last names.


```python
df['Manager'] = df['Manager'].str.strip().str.title().str.replace(r'\s+', ' ', regex=True)
```

```python
df['Manager'].value_counts()
```

Numero de ordenes por ciudad

*Number of orders per city*

```python
df['City'].value_counts()
```

```python
df['Total_Sales'] = df['Price'] * df['Quantity']
```

creamos la columna de ventas totales para identificar cuales son las ciudades, manager y productos que tienen mayores ventas

*Creating the 'Total_Sales' column to identify top-performing cities, managers, and products*

```python
display(df.head())
df.info()
```

Calculamos las ventas totales de todas las ciudades

*Calculating total sales across all cities*

```python
total = df['Total_Sales'].sum()
print(f"El total de ventas fue de: ${total:,.2f}")
```

Buscamos cual de los manager tiene la mayor cantidad de ventas

*Identifying the manager with the highest sales volume*

```python
df.groupby('Manager')['Total_Sales'].sum().sort_values(ascending=False)
```

Agrupamos por ciudad y por Manager para sacar conclusiones y propuestas de mejora

*Grouping by City and Manager to draw conclusions and propose operational improvements*

```python
df.groupby(['Manager', 'City'])['Total_Sales'].agg(['sum', 'mean','count']).style.format("{:,.0f}")
```
